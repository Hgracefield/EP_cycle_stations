```python
import pandas as pd
import numpy as np
import warnings 
import pyarrow.parquet as pq
warnings.filterwarnings('ignore')
```

```python
col = ['ST-1035','ST-454','ST-471']
```

```python
df_2024 = pd.read_parquet('../../Data/sort_data/2024_data.parquet', engine='fastparquet')
df_2024 = df_2024[(df_2024['시작_대여소_ID'].isin(col)) | (df_2024['종료_대여소_ID'].isin(col))]
df_2024.sort_values(['기준_날짜','시간대'],axis=0,ascending=True,inplace=True)
df_2024.reset_index(drop=True, inplace=True)
df_2024.to_csv('../../Data/Zero/2024_data.csv',index=None)
```

```python
df_2025 = pd.read_parquet('../../Data/sort_data/2025_data.parquet', engine='fastparquet')
df_2025 = df_2025[(df_2025['시작_대여소_ID'].isin(col)) | (df_2025['종료_대여소_ID'].isin(col))]
df_2025.sort_values(['기준_날짜','시간대'],axis=0,ascending=True,inplace=True)
df_2025.reset_index(drop=True, inplace=True)
df_2025.to_csv('../../Data/Zero/2025_data.csv',index=None)
```
