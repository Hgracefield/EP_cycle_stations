```python
df = pd.read_csv('../../Data/Zero/Temp/24년도_따릉이_이용건수.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
# df1 = 
df1 = df[(df['전체_이용_분'] >= 5) & (df['전체_이용_거리'] >= 5)]
df1
```

```python
col = ['ST-1035','ST-454','ST-471']

# df_section6 = df1[(df1['시작_대여소_ID'].isin(col) & df1['집계_기준']=='출발시간') | (df1['종료_대여소_ID'].isin(col) & df1['집계_기준']=='도착시간')]
cond1 = (df1['시작_대여소_ID'].isin(col)) & (df1['집계_기준'] == '출발시간')
cond2 = (df1['종료_대여소_ID'].isin(col)) & (df1['집계_기준'] == '도착시간')

df_section6 = df1[cond1 | cond2]
df_section6['시'] = df_section6['기준_시간'] / 100
df_section6['분'] = df_section6['기준_시간'] % 100
df_section6['시'] = df_section6['시'].astype(int)
df_section6['분'] = df_section6['분'].astype(int)
df_section6.sort_values(['기준_날짜','시','분'],ascending=True,inplace=True)
df_section6.reset_index(drop=True, inplace=True)
df_section6
```

```python
import pandas as pd
from glob import glob
station_files = glob('../../Data/Zero/2025/*.csv')
station_files = sorted(station_files)
station_files
```

```python
tmp_raw = []

for file_name in station_files:
    print(file_name)
    try:
        tmp = pd.read_csv(file_name, encoding='euc-kr', encoding_errors='ignore')
    except:
        tmp = pd.read_csv(file_name, encoding='cp949', encoding_errors='ignore')
    tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw,axis=0)
```

```python
station_raw1 = station_raw.iloc[:,9+1:]
```

```python
station_raw1.info()
```

```python
station_raw1['기준_날짜'] = station_raw1['기준_날짜'].astype(int)
station_raw1['전체_건수'] = station_raw1['전체_건수'].astype(int)
station_raw1['전체_이용_분'] = station_raw1['전체_이용_분'].fillna(0)
station_raw1['전체_이용_분'] = station_raw1['전체_이용_분'].astype(int)
station_raw1['전체_이용_거리'] = station_raw1['전체_이용_거리'].fillna(0)
station_raw1['전체_이용_거리'] = station_raw1['전체_이용_거리'].astype(int)
```

```python
station_raw1.to_csv('../../Data/2025_data.csv')
```

```python
import pyarrow as pa
import pyarrow.parquet as pq
```

```python
csv_file = "../../Data/2025.csv"
parquet_file = "../../Data/2025_data.parquet"
chunksize = 500
csv_stream = pd.read_csv(csv_file, sep=',', chunksize=chunksize, low_memory=False)
chunk =next(iter(csv_stream))
```

```python
parquet_schema_old = pa.Table.from_pandas(df=chunk).schema
parquet_schema_new = pa.schema([
    ('기준_날짜', pa.int64()),
    ('집계_기준', pa.string()),
    ('기준_시간대', pa.float64()),
    ('시작_대여소_ID', pa.string()),
    ('시작_대여소명', pa.string()),
    ('종료_대여소_ID', pa.string()),
    ('종료_대여소명', pa.string()),
    ('전체_건수', pa.float64()),
    ('전체_이용_분', pa.float64()),
    ('전체_이용_거리', pa.float64()),
    
])
parquet_schema_old == parquet_schema_new
```

```python
chunksize = 500
csv_stream = pd.read_csv(csv_file, sep=',', chunksize=chunksize, low_memory=False)
for i, chunk in enumerate(csv_stream):
    print("Chunk", i)
    if i == 0:
        # Guess the schema of the CSV file from the first chunk
        parquet_schema = parquet_schema_new
        # Open a Parquet file for writing
        parquet_writer = pq.ParquetWriter(parquet_file, parquet_schema, compression='snappy')
    # Write CSV chunk to the parquet file
    table = pa.Table.from_pandas(chunk, schema=parquet_schema)
    parquet_writer.write_table(table)
else :
    parquet_writer.close()
```
