```python
import pandas as pd
import numpy as np
import warnings 
import pyarrow 
warnings.filterwarnings('ignore')
```

```python
col = ['ST-464','ST-481','ST-479','ST-2247','ST-2252','ST-2425','ST-1027','ST-2263','ST-3126','ST-1035','ST-454','ST-471','ST-1024','ST-1023','ST-1483','ST-1034','ST-462','ST-2264']

df2024 = pd.read_parquet('../../Data/data.parquet')
df2025 = pd.read_parquet('../../Data/2025_data.parquet')
```

```python
total = pd.concat([df2024,df2025],axis=0)
total
```

```python
cond1 = (total['시작_대여소_ID'].isin(col)) & (total['집계_기준'] == '출발시간')
cond2 = (total['종료_대여소_ID'].isin(col)) & (total['집계_기준'] == '도착시간')

result = total[cond1 | cond2]
# df_section6['시'] = df_section6['기준_시간'] / 100
# df_section6['분'] = df_section6['기준_시간'] % 100
# df_section6['시'] = df_section6['시'].astype(int)
# df_section6['분'] = df_section6['분'].astype(int)
result.sort_values(['기준_날짜'],ascending=True,inplace=True)
result.reset_index(drop=True, inplace=True)
result
```

```python
result.to_csv('../../Data/total_data.csv',index=None)
```
