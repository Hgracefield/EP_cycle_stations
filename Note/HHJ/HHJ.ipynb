```python
import pandas as pd

group_cols = ['요일', '월', '주말', '집계_기준', '기준_시간대', '종료_대여소_ID']

def make_grouped(df):
    df = df.copy()

    df['기준_날짜'] = pd.to_datetime(df['기준_날짜'].astype(str), format='%Y%m%d')
    df['요일'] = df['기준_날짜'].dt.day_name()
    df['월'] = df['기준_날짜'].dt.month
    df['주말'] = df['기준_날짜'].dt.weekday >= 5

    grouped = df.groupby(group_cols, as_index=False).agg({
        '전체_건수': 'sum',
        '전체_이용_분': 'sum',
        '전체_이용_거리': 'sum'
    })

    return grouped

st_1024_grouped = make_grouped(st_1024)
st_1023_grouped = make_grouped(st_1023)
st_1483_grouped = make_grouped(st_1483)
```

> 누적의 최대값을 구하기 누적(출발 - 도착)             
출발, 도착, 시간순 누적했을때 가장 크게 부족해지는 순간의 값을 필요_보충_대수 로 두기.              
새벽 4시 보충, 오후 4시 보충

```python
import pandas as pd
import numpy as np

# -----------------------------
# 1. 원본 불러오기
# -----------------------------
df = pd.read_csv("../../Data/ST-1024.csv")

# -----------------------------
# 2. 기본 전처리
# -----------------------------
# 날짜형 변환
df['기준_날짜'] = pd.to_datetime(df['기준_날짜'].astype(str), format='%Y%m%d')

# 시간대 숫자 정리
df['기준_시간대'] = df['기준_시간대'].fillna(-1).astype(int)

# HHMM -> 시, 분 분리
df['hour'] = df['기준_시간대'] // 100
df['minute'] = df['기준_시간대'] % 100

# datetime 만들기
df['datetime'] = df['기준_날짜'] + pd.to_timedelta(df['hour'], unit='h') + pd.to_timedelta(df['minute'], unit='m')

# 필요 없는 완전 이상치 제거:
# 전체 이용분, 전체 이용거리가 둘 다 0인 행만 제거
df = df[
    ~(
        (df['전체_이용_분'] == 0) &
        (df['전체_이용_거리'] == 0)
    )
].copy()

# 출발/도착 건수 통일
# 출발시간이면 -건수, 도착시간이면 +건수
df['signed_count'] = np.where(
    df['집계_기준'] == '출발시간',
    -df['전체_건수'],
    df['전체_건수']
)

# -----------------------------
# 3. 보충 구간(새벽4시 / 오후4시) 라벨링
# -----------------------------
def assign_restock_block(dt):
    h = dt.hour
    
    # 04:00 ~ 15:59  -> 새벽4시 보충 블록
    if 4 <= h < 16:
        block_start = dt.normalize() + pd.Timedelta(hours=4)
        block_name = '새벽4시보충'
    else:
        # 16:00 ~ 23:59 -> 당일 오후4시 시작
        # 00:00 ~ 03:59 -> 전날 오후4시 시작
        if h >= 16:
            block_start = dt.normalize() + pd.Timedelta(hours=16)
        else:
            block_start = (dt.normalize() - pd.Timedelta(days=1)) + pd.Timedelta(hours=16)
        block_name = '오후4시보충'
    
    return pd.Series([block_name, block_start])

df[['보충타입', '보충시작시각']] = df['datetime'].apply(assign_restock_block)

# -----------------------------
# 4. 블록 내부 시간순 정렬
# -----------------------------
df = df.sort_values(['보충시작시각', 'datetime']).copy()

# -----------------------------
# 5. 블록별 타깃 만들기
#    필요_보충_대수 = max(누적 출발-도착 부족량)
# -----------------------------
def make_target(group):
    # 출발은 -, 도착은 + 이므로
    # 자전거 부족량 관점에서는 -(signed_count)를 누적
    # 출발 발생 시 부족 증가, 도착 발생 시 부족 감소
    shortage_flow = -group['signed_count']
    cumulative_shortage = shortage_flow.cumsum()
    
    need_bikes = cumulative_shortage.max()
    net_shortage = shortage_flow.sum()  # 총 순부족
    
    return pd.Series({
        '필요_보충_대수': max(0, need_bikes),
        '순부족_대수': net_shortage,
        '총출발건수': group.loc[group['집계_기준'] == '출발시간', '전체_건수'].sum(),
        '총도착건수': group.loc[group['집계_기준'] == '도착시간', '전체_건수'].sum(),
        '레코드수': len(group)
    })

target_df = df.groupby(['보충시작시각', '보충타입']).apply(make_target).reset_index()

# -----------------------------
# 6. 날짜 파생 변수 만들기
# -----------------------------
target_df['year'] = target_df['보충시작시각'].dt.year
target_df['month'] = target_df['보충시작시각'].dt.month
target_df['day'] = target_df['보충시작시각'].dt.day
target_df['weekday'] = target_df['보충시작시각'].dt.weekday   # 월=0, 일=6
target_df['is_weekend'] = target_df['weekday'].isin([5, 6]).astype(int)
target_df['hour_block'] = target_df['보충시작시각'].dt.hour   # 4 or 16

# -----------------------------
# 7. 순환형 인코딩(월, 요일)
# -----------------------------
target_df['month_sin'] = np.sin(2 * np.pi * target_df['month'] / 12)
target_df['month_cos'] = np.cos(2 * np.pi * target_df['month'] / 12)

target_df['weekday_sin'] = np.sin(2 * np.pi * target_df['weekday'] / 7)
target_df['weekday_cos'] = np.cos(2 * np.pi * target_df['weekday'] / 7)

# -----------------------------
# 8. 직전 블록 정보(lag feature)
# -----------------------------
target_df = target_df.sort_values('보충시작시각').copy()

target_df['prev_need_bikes'] = target_df['필요_보충_대수'].shift(1)
target_df['prev_net_shortage'] = target_df['순부족_대수'].shift(1)
target_df['prev_departures'] = target_df['총출발건수'].shift(1)
target_df['prev_arrivals'] = target_df['총도착건수'].shift(1)

# 최근 3개 블록 평균
target_df['rolling_need_3'] = target_df['필요_보충_대수'].shift(1).rolling(3).mean()
target_df['rolling_dep_3'] = target_df['총출발건수'].shift(1).rolling(3).mean()
target_df['rolling_arr_3'] = target_df['총도착건수'].shift(1).rolling(3).mean()

# -----------------------------
# 9. 모델 입력용 최종 데이터
# -----------------------------
model_df = target_df.dropna().copy()

feature_cols = [
    'month',
    'day',
    'weekday',
    'is_weekend',
    'hour_block',
    'month_sin',
    'month_cos',
    'weekday_sin',
    'weekday_cos',
    'prev_need_bikes',
    'prev_net_shortage',
    'prev_departures',
    'prev_arrivals',
    'rolling_need_3',
    'rolling_dep_3',
    'rolling_arr_3'
]

X = model_df[feature_cols]
y = model_df['필요_보충_대수']

print(model_df.head())
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error,  r2_score

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

model = RandomForestRegressor(
    n_estimators=300,
    max_depth=8,
    random_state=42
)

model.fit(X_train, y_train)
pred = model.predict(X_test)
r2 = r2_score(y_test, pred)



mae = mean_absolute_error(y_test, pred)
rmse = mean_squared_error(y_test, pred) ** 0.5

print("MAE:", mae) # 평균 절대 오차
print("RMSE:", rmse)  # 오차를 제곱해서 평균낸 뒤 다시 루트를 씌운다 -> 큰 오차에 더 큰 패널티를 주는 지표
print("R2:", r2)
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from xgboost import XGBRegressor

# -----------------------------
# 1. train / test 분리
# -----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False
)

# -----------------------------
# 2. 모델 정의
# -----------------------------
models = {
    "RandomForest": RandomForestRegressor(
        n_estimators=300,
        max_depth=8,
        random_state=42
    ),

    "XGBoost": XGBRegressor(
        n_estimators=300,
        learning_rate=0.05,
        max_depth=4,
        random_state=42
    ),

    "SVM": Pipeline([
        ("scaler", StandardScaler()),
        ("model", SVR(
            kernel='rbf',
            C=10,
            epsilon=0.1
        ))
    ]),

    "KNN": Pipeline([
        ("scaler", StandardScaler()),
        ("model", KNeighborsRegressor(
            n_neighbors=5
        ))
    ])
}

# -----------------------------
# 3. 학습 및 평가
# -----------------------------
results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, pred)
    rmse = mean_squared_error(y_test, pred) ** 0.53

    results.append({
        "Model": name,
        "MAE": mae,
        "RMSE": rmse
    })

    print(f"[{name}]")
    print("MAE :", mae)
    print("RMSE:", rmse)
    print("-" * 30)

# -----------------------------
# 4. 결과표 정리
# -----------------------------
result_df = pd.DataFrame(results).sort_values("RMSE")
print(result_df)
```

> 평균 오차만 보면 KNN이 가장 낮았지만, 큰 오차까지 반영하는 RMSE는 RandomForest가 가장 낮았다. 실제 자전거 보충 예측에서는 일부 시점의 큰 오차를 줄이는 것이 중요하므로, 최종적으로 RandomForest를 더 안정적인 모델로 판단하였다.

```python
# 함수
import pandas as pd
import numpy as np
import os

# -----------------------------
# 1. 대여소 1개 파일 처리 함수
# -----------------------------
def make_station_model_df(file_path, station_name=None):
    df = pd.read_csv(file_path)

    # 파일명으로 대여소명 자동 추출
    if station_name is None:
        station_name = os.path.splitext(os.path.basename(file_path))[0]

    # -----------------------------
    # 2. 기본 전처리
    # -----------------------------
    df['기준_날짜'] = pd.to_datetime(df['기준_날짜'].astype(str), format='%Y%m%d')
    df['기준_시간대'] = df['기준_시간대'].fillna(-1).astype(int)

    df['hour'] = df['기준_시간대'] // 100
    df['minute'] = df['기준_시간대'] % 100

    df['datetime'] = (
        df['기준_날짜']
        + pd.to_timedelta(df['hour'], unit='h')
        + pd.to_timedelta(df['minute'], unit='m')
    )

    # 둘 다 0인 행만 제거
    df = df[
        ~(
            (df['전체_이용_분'] == 0) &
            (df['전체_이용_거리'] == 0)
        )
    ].copy()

    # 출발/도착 부호 통일
    df['signed_count'] = np.where(
        df['집계_기준'] == '출발시간',
        -df['전체_건수'],
        df['전체_건수']
    )

    # -----------------------------
    # 3. 보충 구간 라벨링
    # -----------------------------
    def assign_restock_block(dt):
        h = dt.hour

        if 4 <= h < 16:
            block_start = dt.normalize() + pd.Timedelta(hours=4)
            block_name = '새벽4시보충'
        else:
            if h >= 16:
                block_start = dt.normalize() + pd.Timedelta(hours=16)
            else:
                block_start = (dt.normalize() - pd.Timedelta(days=1)) + pd.Timedelta(hours=16)
            block_name = '오후4시보충'

        return pd.Series([block_name, block_start])

    df[['보충타입', '보충시작시각']] = df['datetime'].apply(assign_restock_block)

    # 시간순 정렬
    df = df.sort_values(['보충시작시각', 'datetime']).copy()

    # -----------------------------
    # 4. 블록별 타깃 만들기
    # -----------------------------
    def make_target(group):
        shortage_flow = -group['signed_count']
        cumulative_shortage = shortage_flow.cumsum()

        need_bikes = cumulative_shortage.max()
        net_shortage = shortage_flow.sum()

        return pd.Series({
            '필요_보충_대수': max(0, need_bikes),
            '순부족_대수': net_shortage,
            '총출발건수': group.loc[group['집계_기준'] == '출발시간', '전체_건수'].sum(),
            '총도착건수': group.loc[group['집계_기준'] == '도착시간', '전체_건수'].sum(),
            '레코드수': len(group)
        })

    target_df = df.groupby(['보충시작시각', '보충타입']).apply(make_target).reset_index()

    # -----------------------------
    # 5. 날짜 파생 변수
    # -----------------------------
    target_df['year'] = target_df['보충시작시각'].dt.year
    target_df['month'] = target_df['보충시작시각'].dt.month
    target_df['day'] = target_df['보충시작시각'].dt.day
    target_df['weekday'] = target_df['보충시작시각'].dt.weekday
    target_df['is_weekend'] = target_df['weekday'].isin([5, 6]).astype(int)
    target_df['hour_block'] = target_df['보충시작시각'].dt.hour

    # 순환형 인코딩
    target_df['month_sin'] = np.sin(2 * np.pi * target_df['month'] / 12)
    target_df['month_cos'] = np.cos(2 * np.pi * target_df['month'] / 12)

    target_df['weekday_sin'] = np.sin(2 * np.pi * target_df['weekday'] / 7)
    target_df['weekday_cos'] = np.cos(2 * np.pi * target_df['weekday'] / 7)

    # -----------------------------
    # 6. lag feature
    # -----------------------------
    target_df = target_df.sort_values('보충시작시각').copy()

    target_df['prev_need_bikes'] = target_df['필요_보충_대수'].shift(1)
    target_df['prev_net_shortage'] = target_df['순부족_대수'].shift(1)
    target_df['prev_departures'] = target_df['총출발건수'].shift(1)
    target_df['prev_arrivals'] = target_df['총도착건수'].shift(1)

    target_df['rolling_need_3'] = target_df['필요_보충_대수'].shift(1).rolling(3).mean()
    target_df['rolling_dep_3'] = target_df['총출발건수'].shift(1).rolling(3).mean()
    target_df['rolling_arr_3'] = target_df['총도착건수'].shift(1).rolling(3).mean()

    # 대여소 구분 컬럼 추가
    target_df['station'] = station_name

    # -----------------------------
    # 7. 모델 입력용 최종 데이터
    # -----------------------------
    model_df = target_df.dropna().copy()

    feature_cols = [
        'month',
        'day',
        'weekday',
        'is_weekend',
        'hour_block',
        'month_sin',
        'month_cos',
        'weekday_sin',
        'weekday_cos',
        'prev_need_bikes',
        'prev_net_shortage',
        'prev_departures',
        'prev_arrivals',
        'rolling_need_3',
        'rolling_dep_3',
        'rolling_arr_3'
    ]

    X = model_df[feature_cols]
    y = model_df['필요_보충_대수']

    return {
        'station': station_name,
        'raw_df': df,
        'target_df': target_df,
        'model_df': model_df,
        'X': X,
        'y': y
    }
```

```python
file_paths = [
    "../../Data/ST-1024.csv",
    "../../Data/ST-1023.csv",
    "../../Data/ST-1483.csv"
]

results = {}

for path in file_paths:
    result = make_station_model_df(path)
    results[result['station']] = result

# 확인
for station, result in results.items():
    print(f"[{station}]")
    print("model_df shape:", result['model_df'].shape)
    print("X shape:", result['X'].shape)
    print("y shape:", result['y'].shape)
    print("-" * 40)
```

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error

from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

from xgboost import XGBRegressor


def evaluate_models(X, y, test_size=0.2, random_state=42, verbose=True):
    # -----------------------------
    # 1. train / test 분리
    # -----------------------------
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, shuffle=False
    )

    # -----------------------------
    # 2. 모델 정의
    # -----------------------------
    models = {
        "RandomForest": RandomForestRegressor(
            n_estimators=300,
            max_depth=8,
            random_state=random_state
        ),

        "XGBoost": XGBRegressor(
            n_estimators=300,
            learning_rate=0.05,
            max_depth=4,
            random_state=random_state
        ),

        "SVM": Pipeline([
            ("scaler", StandardScaler()),
            ("model", SVR(
                kernel='rbf',
                C=10,
                epsilon=0.1
            ))
        ]),

        "KNN": Pipeline([
            ("scaler", StandardScaler()),
            ("model", KNeighborsRegressor(
                n_neighbors=5
            ))
        ])
    }

    # -----------------------------
    # 3. 학습 및 평가
    # -----------------------------
    results = []
    trained_models = {}

    for name, model in models.items():
        model.fit(X_train, y_train)
        pred = model.predict(X_test)

        mae = mean_absolute_error(y_test, pred)
        rmse = mean_squared_error(y_test, pred) ** 0.5

        results.append({
            "Model": name,
            "MAE": mae,
            "RMSE": rmse
        })

        trained_models[name] = model

        if verbose:
            print(f"[{name}]")
            print("MAE :", mae)
            print("RMSE:", rmse)
            print("-" * 30)

    # -----------------------------
    # 4. 결과표 정리
    # -----------------------------
    result_df = pd.DataFrame(results).sort_values("RMSE").reset_index(drop=True)

    return {
        "result_df": result_df,
        "trained_models": trained_models,
        "X_train": X_train,
        "X_test": X_test,
        "y_train": y_train,
        "y_test": y_test
    }
```

```python
results = {}

for path in [
    "../../Data/ST-1024.csv",
    "../../Data/ST-1023.csv",
    "../../Data/ST-1483.csv"
]:
    result = make_station_model_df(path)
    results[result['station']] = result
```

```python
station_eval_results = {}

for station in ['ST-1024', 'ST-1023', 'ST-1483']:
    print(f"\n===== {station} =====")
    station_eval_results[station] = evaluate_models(
        results[station]['X'],
        results[station]['y']
    )
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_model_performance(result_df, title='Model Performance Comparison'):
    x = np.arange(len(result_df['Model']))
    width = 0.35

    fig, ax = plt.subplots(figsize=(9, 5))
    
    bars1 = ax.bar(
        x - width/2,
        result_df['MAE'],
        width,
        label='MAE',
        edgecolor='black'
    )
    
    bars2 = ax.bar(
        x + width/2,
        result_df['RMSE'],
        width,
        label='RMSE',
        edgecolor='black'
    )

    ax.set_xticks(x)
    ax.set_xticklabels(result_df['Model'])
    ax.set_title(title)
    ax.set_xlabel('Model')
    ax.set_ylabel('Error')
    ax.legend()
    ax.grid(axis='y', linestyle='--', alpha=0.6)

    # 막대 위 숫자 표시
    for bar in bars1:
        height = bar.get_height()
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            height + 0.01,
            f'{height:.3f}',
            ha='center',
            va='bottom',
            fontsize=9
        )

    for bar in bars2:
        height = bar.get_height()
        ax.text(
            bar.get_x() + bar.get_width() / 2,
            height + 0.01,
            f'{height:.3f}',
            ha='center',
            va='bottom',
            fontsize=9
        )

    plt.tight_layout()
    plt.show()
```
